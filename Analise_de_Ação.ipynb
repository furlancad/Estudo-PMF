```python
vale_open_close = vale.drop(['Open','High', 'Low','Adj Close','Volume'], axis=1)
vale_open_close
```

```python
vale_open_close['Valor anterior'] = vale_open_close['Close'].shift()
vale_open_close
```

```python
vale_open_close['Valor futuro'] = vale_open_close['Close'].shift(-1)
vale_open_close
```

```python
vale_open_close['Rentabilidade'] = vale_open_close['Close'].pct_change()*100
vale_open_close
```

```python
plt.plot(vale['Close'].loc['2023-01-01':], 'go-', label='line 1', linewidth=4)
plt.plot(petr['Close'].loc['2023-01-01':], 'rs-', label='line 2',)

plt.legend(["VALE3", "PETR4"])
plt.grid(True)

# Salve o gráfico como uma imagem
fig = plt.gcf()
fig.savefig("image.png")
```

```python
plt.plot(vale['Close'].loc['2016-01-01':], 'g-', label='line 1', linewidth=2)
plt.plot(vale.Close.rolling(50).mean().loc['2016-01-01':], 'b-', label='line 2',)
plt.plot(vale.Close.rolling(200).mean().loc['2016-01-01':], 'r-', label='line 2',)

plt.legend(["VALE3", "moving_average_50", 'moving_average_200'])
plt.grid(True)

# Salve o gráfico como uma imagem
fig = plt.gcf()
fig.savefig("image.png")
```
